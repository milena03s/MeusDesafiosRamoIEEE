```python

!pip install requests
!pip install beautifulsoup4

# Importar as bibliotecas
import requests
from bs4 import BeautifulSoup
import os
import datetime

# Criação de pasta
def criar_pasta(nome_pasta):
    try:
        os.mkdir(nome_pasta)
    except FileExistsError:
        print(f"A pasta '{nome_pasta}' já existe!")

# Baixar imagens
def baixar_imagens(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # achar todas as tags img no html
    imagens = soup.find_all('img')
    print(f"Total de imagens encontradas: {len(imagens)}")

    #diretório fica com nome da data atual
    nome_pasta = datetime.datetime.today().strftime("%Y-%m-%d")
    criar_pasta(nome_pasta)

    for i, img in enumerate(imagens):
        # Pega o URL do src, adiciona https onde precisa pra não dar erro
        src = img.get('src')
        if src:
            if not src.startswith('http'):
                src ="https:"+src

            # Baixa e salva a imagem
        try:
                img_baixar = requests.get(src).content
                with open(f"{nome_pasta}/imagem_{i+1}.jpg", 'wb') as f:
                    f.write(img_baixar)
                print(f"Imagem {i+1} baixada com sucesso!")
        except Exception as e:
                print(f"Erro ao baixar a imagem! {e}")
# URL de exemplo
url = input("Insira a URL do site: ")
baixar_imagens(url)
```
